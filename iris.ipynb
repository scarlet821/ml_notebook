```python
from sklearn import datasets
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(iris['data'])
df = df.rename(
    columns={
        0: 'sepal_l',
        1: 'sepal_w',
        2: 'petal_l',
        3: 'petal_w'
    }
)

df.head()
df.corr()
```

```python
df.head()
```

```python
sns.heatmap(df.corr(), cmap='coolwarm', annot=True)
```

```python
iris.feature_names
```

```python
iris.target
print(len(iris.target), len(df))
```

```python
df['species'] = iris.target
```

```python
sns.lmplot(x="sepal_l", y="petal_l", hue="species", data = df, \
              palette="Set1", markers=["o", "s", "D"])

```

```python
sns.lmplot(x="sepal_l", y="petal_l", hue="species", data = df, \
              palette="Set1", markers=["o", "s", "D"], fit_reg=False)

```

```python
sns.pairplot(vars=['sepal_l', 'petal_l'], hue='species', markers=['o', 's', 'D'], data=df)
```

```python
sns.swarmplot(x="species", y="petal_l", data=df)
```
