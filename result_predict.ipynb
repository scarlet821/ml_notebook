```python
features1 = ['exercise_frequency','study_hours_per_day', 'mental_health_rating','sleep_hours' , 'parental_education_level']
features2 = ['study_hours_per_day', 'attendance_percentage', 'mental_health_rating', 'sleep_hours', 'part_time_job']
target = ['exam_score']
```

```python
df_model1 = df[features1 + target].copy()
df_model2 = df[features2 + target].copy()
df_model1.head()
```

```python
df_model2.head()
```

```python
le1 = LabelEncoder()
le2 = LabelEncoder()
```

```python
df_model1['parental_education_level'] = le1.fit_transform(df['parental_education_level'])

df_model1.head()
```

```python
df_model2['part_time_job'] = le2.fit_transform(df['part_time_job'])

df_model2.head()
```

```python
X1 = df_model1[features1]
X2 = df_model2[features2]

y = df_model1[target]
```

```python
X_train1, X_test1, y_train, y_test = train_test_split(X1,y, test_size=0.2)
X_train2, X_test2, y_train, y_test = train_test_split(X2,y, test_size=0.2)
```
